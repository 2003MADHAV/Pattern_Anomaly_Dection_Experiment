```python
import matplotlib.pyplot as plt

errors_data = [3, 7, 8, 5, 12, 14, 21, 13, 18]

plt.boxplot(errors_data)
plt.title('Box-and-Whisker Plot for server-logs')
plt.ylabel('Values')
plt.show()
```

```python
import matplotlib.pyplot as plt

data = [3, 7, 8, 5, 12, 14, 21, 13, 18]

plt.boxplot(data, showfliers=True)
plt.title('Box-and-Whisker Plot with Outliers')
plt.ylabel('Values')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = [3, 7, 8, 5, 12, 14, 21, 13, 18]

# Create a box-and-whisker plot with stripplot
sns.boxplot(x=data)
sns.stripplot(x=data, color='black', size=8, jitter=True, alpha=0.7)

plt.title('Box-and-Whisker Plot with Individual Data Points')
plt.xlabel('Values')
plt.show()
```
