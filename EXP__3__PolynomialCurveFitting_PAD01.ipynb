```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python

np.random.seed(42)
months = np.arange(1, 13, 1)
historical_temperatures = 25 + 5 * np.sin(0.5 * np.pi * months) + np.random.normal(0, 2, 12)
```

```python

months = months.reshape(-1, 1)
historical_temperatures = historical_temperatures.reshape(-1, 1)
```

```python

linear_model = LinearRegression()
linear_model.fit(months, historical_temperatures)
linear_predictions = linear_model.predict(months)
```

```python

quadratic_features = PolynomialFeatures(degree=2)
X_quadratic = quadratic_features.fit_transform(months)
quadratic_model = LinearRegression()
quadratic_model.fit(X_quadratic, historical_temperatures)
quadratic_predictions = quadratic_model.predict(X_quadratic)
```

```python

cubic_features = PolynomialFeatures(degree=3)
X_cubic = cubic_features.fit_transform(months)
cubic_model = LinearRegression()
cubic_model.fit(X_cubic, historical_temperatures)
cubic_predictions = cubic_model.predict(X_cubic)
```

```python

plt.scatter(months, historical_temperatures, label='Actual Data')
plt.plot(months, linear_predictions, label='Linear Fit', color='red')
plt.plot(months, quadratic_predictions, label='Quadratic Fit', color='green')
plt.plot(months, cubic_predictions, label='Cubic Fit', color='blue')
plt.title('Polynomial Curve Fitting')
plt.xlabel('Months')
plt.ylabel('Temperature')
plt.legend()
plt.show()
```

```python

mse_linear = mean_squared_error(historical_temperatures, linear_predictions)
mse_quadratic = mean_squared_error(historical_temperatures, quadratic_predictions)
mse_cubic = mean_squared_error(historical_temperatures, cubic_predictions)
```

```python
print(f'Mean Squared Error (Linear Fit): {mse_linear:.2f}')
print(f'Mean Squared Error (Quadratic Fit): {mse_quadratic:.2f}')
print(f'Mean Squared Error (Cubic Fit): {mse_cubic:.2f}')
```
