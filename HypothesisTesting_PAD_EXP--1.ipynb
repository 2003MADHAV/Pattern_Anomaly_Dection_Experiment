```python
import scipy.stats as stats
# Sample data (replace with your actual data)
treatment_group = [150, 155, 160, 165, 170]
control_group = [140, 145, 150, 155, 160]
# Perform two-sample t-test
t_stat, p_value = stats.ttest_ind(treatment_group, control_group)
# Output results
print("T-statistic:", t_stat)
print("P-value:", p_value)
```

```python
alpha = 0.05
if p_value < alpha:
  print("Reject the null hypothesis. There is evidence of a significant difference.")
else:
  print("Fail to reject the null hypothesis. The difference is not statistically significant.")
```

```python
import numpy as np
# Sample data (replace with your actual data)
sample_data = [160, 165, 170, 175, 180]
# Assumed population mean (replace with your actual population mean)
population_mean = 160
# Perform one-sample z-test
z_stat = (np.mean(sample_data) - population_mean) / (np.std(sample_data) / np.sqrt(len(sample_data)))
# P-value can be obtained using a standard normal distribution table or a statistical library
p_value = stats.norm.cdf(z_stat)
# Output results
print("Z-statistic:", z_stat)
print("P-value:", p_value)
```

```python
alpha = 0.05
if p_value < alpha:
  print("Reject the null hypothesis. There is evidence of a significant difference.")
else:
  print("Fail to reject the null hypothesis. The difference is not statistically significant.")
```
