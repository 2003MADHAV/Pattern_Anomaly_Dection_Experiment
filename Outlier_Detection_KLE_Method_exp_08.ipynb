```python
X=data.iloc[:,:]

import numpy as np

def euclidean_distance(point1,point2):
  return np.sqrt(np.sum((point1-point2)**2))

def calculate_local_density(point, dataset,threshold):
  count=0
  for other_point in dataset:
    if euclidean_distance(point,other_point)<=threshold:
      count +=1
    return count
def calculate_kle_score(point, dataset, threshold):
    local_density = calculate_local_density(point, dataset, threshold)
    sum_avg_density = 0  # Fix: Correct the assignment here
    for other_point in dataset:
        if not np.array_equal(other_point, point):
            other_point_density = calculate_local_density(other_point, dataset, threshold)
            sum_avg_density += other_point_density

    # Calculate the average density of neighboring points
    avg_density = sum_avg_density / (len(dataset) - 1)  # -1 to exclude the point itself

    # Calculate KLE score
    kle_score = local_density / (local_density + avg_density)

    return kle_score

def detect_outliers(dataset, threshold, kle_threshold):
    outliers = []
    outlier_scores = {}
    for i, point in enumerate(dataset):
        kle_score = calculate_kle_score(point, dataset, threshold)
        if kle_score > kle_threshold:
            outliers.append(i)
        outlier_scores[i] = kle_score

    return outliers, outlier_scores

# Example dataset
dataset = np.random.rand(100, 11)

# Set threshold and KLE threshold
threshold = 0.5
kle_threshold = 0.8

# Detect outliers and get outlier scores
outliers, outlier_scores = detect_outliers(dataset, threshold, kle_threshold)

# Print outlier scores of all points
print("Outlier scores of all points:")
for point, score in outlier_scores.items():
    print(f"Point: {point}, KLE Score: {score}")
```
